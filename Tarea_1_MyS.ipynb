##Ejemplo 1
metodo de euler con 
\begin{equation}
\frac{dy}{dt} = ln(2)y(t)
\end{equation}

con $y(0)=1$

```
def step_euler(y,h,f):
  next_y=y+h*f(y)
  return next_y

def euler(h,f,y_0,start_t,end_t):
  N= int((end_t - start_t)/h)
  t_list=np.linspace(start_t,end_t,N+1)
  y_list=np.zeros(N+1)
  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1]=step_euler(y_list[i],h,f)
  return t_list,y_list
```

```
def g(y):
  crecimiento=np.log(2)*y
  return crecimiento


y_0=1
h=0.01
t_0=0
t_N=10

t_list,y_list=euler(h,g,y_0,t_0,t_N)

plt.plot(t_list,y_list)
plt.plot(t_list, np.power(2,t_list))
plt.grid()
```

Realice el ejercicio anterior utilizando el metodo de euler mejorado

```
def step_euler_mejorado(y,h,f):
  next_y = y+0.5*(f(y)+f(step_euler(y,h,f)))*h
  return next_y

def euler_mejorado(h, f, y_0, t_i, t_f):
  N = int((t_f - t_i)/h)
  t_list = np.linspace(t_i, t_f, N+1)
  y_list = np.zeros(N+1)
  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1] = step_euler_mejorado(y_list[i], h, f)
  return t_list, y_list
```

```
y_0 = 1
h=[1,0.1,0.01,0.001,0.0001]
t_0=0
t_N=1
plt.figure(figsize=(7,6))
for i in h :
  t_list,y_list=euler_mejorado(i,g,y_0,t_0,t_N)
  plt.plot(t_list,y_list)
plt.grid()
```

```

def rk4(f,x0,t0,tN,N):
  t= np.linspace(t0,tN,N+1)
  h= (tN-t0)/N
  x_4rk=np.zeros(N+1)
  x_4rk[0]= x0                         # Condicion inicial

  for n in range(N):
      k1 = h*f( x_4rk[n]       , t[n]         )
      k2 = h*f( x_4rk[n] + k1/2, t[n] + (h/2) ) 
      k3 = h*f( x_4rk[n] + k2/2, t[n] + (h/2) ) 
      k4 = h*f( x_4rk[n] + k3  , t[n] +  h    )
      
      x_4rk[n+1] = x_4rk[n] + k1/6 + k2/3 + k3/3 + k4/6
  return t,x_4rk
```
