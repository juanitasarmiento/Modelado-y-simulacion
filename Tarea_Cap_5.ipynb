El amor de romeo sera proporcional al amor de julieta, en cambio el amor de julieta sera inverso al amor de romeo.



```python
A=np.array([[0,1],
           [-1,1]])
delta=np.linalg.det(A)
traza=np.trace(A)
L,V=np.linalg.eig(A)
print('valores propios',L,'\ntraza =',traza,'\ndeterminante =',delta)
```

La traza y el determinante son possitivos lo que implica que la clasificacion del pinto fijo una espiral inestable.

```python
X,Y=np.meshgrid(np.linspace(-15,15,25),np.linspace(-15,15,25))
V= -X+Y
U= Y
fig, ax = plt.subplots(figsize=(12,12))
Q= ax.quiver(X,Y,U,V,units ='width')

def vdp(t,z):
  x,y=z 
  #Donde, j=y
  return[y,-x+y] #Ecuaciones del sistema

ti,tf=0,7
t=np.linspace(ti,tf,500)

for x0 in range(-2 , 3):
  for y0 in range(-2 , 3):
    for style in zip(styles):
      sol=solve_ivp(vdp,[ti,tf],[x0,y0],t_eval=t)
      plt.plot(sol.y[0],sol.y[1],style)
plt.axis([-15,15,-15,15])
plt.grid()
```

### 5.3.3
(fuera de contacto con sus propios sentimientos) suponga que romeo y julieta reaccionan el uno al otro, pero no a ellos mismos: $\dot{R}=aJ$, $\dot{J}=bR$ ¿qué pasaria?

```python
a = sp.Symbol('a')
b = sp.Symbol('b')
A=sp.Matrix([[0,a],
           [b,0]])
delta=sp.det(A)
traza=sp.trace(A)
L=A.eigenvals()
V=A.eigenvects()
print('valores propios',L,'\ntraza =',traza,'\ndeterminante =',delta)
```

Con traza 0 obtenemos centros neutrales y solo variara el determinante dependiendo de los valores de a y b, cuando el determinante es menor obtenemos un punto de silla, cuando es mayor optenemos elipces.


```python
for a in [-1,1]:
  for b in [-1,1]:
    A=np.array([[0,a],
              [b,0]])
    delta=np.linalg.det(A)
    traza=np.trace(A)
    print('para a=',a,'y b =',b,'Con traza =',traza,'y con determinante =',delta)
    X,Y=np.meshgrid(np.linspace(-2,2,12),np.linspace(-2,2,12))
    V= b*X
    U= a*Y
    fig, ax = plt.subplots()
    Q= ax.quiver(X,Y,U,V,units ='width')

    def vdp(t,z):
      x,y=z 
      #Donde, j=y
      return[a*y,b*x] #Ecuaciones del sistema

    ti,tf=0,10
    t=np.linspace(ti,tf,500)
    for x0 in range(-2 , 3):
      for y0 in range(-2 , 3):
        for style in zip(styles):
          sol=solve_ivp(vdp,[ti,tf],[x0,y0],t_eval=t)
          plt.plot(sol.y[0],sol.y[1],style)
        plt.axis([-2,2,-2,2])
        plt.grid()
    ax.spines['left'].set_position(('data',0.0))
    ax.spines['bottom'].set_position(('data',0.0))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    plt.show()
```

### 5.3.4
(Fuego y agua) ¿Los opuestos se atraen? Analizar $\dot{R}=aR+bJ$ , $\dot{J}=-bR-aJ$

```python
a = sp.Symbol('a')
b = sp.Symbol('b')
A=sp.Matrix([[a,b],
           [-b,-a]])
delta=sp.det(A)
traza=sp.trace(A)
L=A.eigenvals()
V=A.eigenvects()
print('valores propios',L,'\ntraza =',traza,'\ndeterminante =',delta)
```

Con traza igual a 0 obtenemos centros neutrales, cuando el determinante es menor a 0 son puntos de silla y al ser mayor a 0 obtenemos elipces, tambien al tener un determinante igual a 0 obtenemos lineas de puntos fijos.

```python
for a in [1,2,3]:
    b = 2
    A=np.array([[a,b],
              [-b,-a]])
    delta=np.linalg.det(A)
    traza=np.trace(A)
    print('para a=',a,'y b =',b,'Con traza =',traza,'y con determinante =',delta)
    X,Y=np.meshgrid(np.linspace(-2,2,12),np.linspace(-2,2,12))
    V= -b*X-a*Y
    U= a*X+ b*Y
    fig, ax = plt.subplots()
    Q= ax.quiver(X,Y,U,V,units ='width')

    def vdp(t,z):
      x,y=z 
      #Donde, j=y
      return[a*x+ b*y,-b*x-a*y] #Ecuaciones del sistema

    ti,tf=0,10
    t=np.linspace(ti,tf,500)
    for x0 in range(-4 , 3):
      for y0 in range(-4 , 3):
        for style in zip(styles):
          sol=solve_ivp(vdp,[ti,tf],[x0,y0],t_eval=t)
          plt.plot(sol.y[0],sol.y[1],style)
        plt.axis([-2,2,-2,2])
        plt.grid()
    ax.spines['left'].set_position(('data',0.0))
    ax.spines['bottom'].set_position(('data',0.0))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    plt.show()
```

```python
a = sp.Symbol('a')
b = sp.Symbol('b')
A=sp.Matrix([[a,b],
           [b,a]])
delta=sp.det(A)
traza=sp.trace(A)
L=A.eigenvals()
V=A.eigenvects()
print('valores propios',L,'\ntraza =',traza,'\ndeterminante =',delta)
```

Con determinante negativo obtenemos puntos de silla el cual sera estable con una traza negativa e inestable con una positiva.

Con determinante positivo en este caso obtenemos nodos los cuales seran unestables con traza negativa e inestables con traza positiva.

Y con determinantes igual a 0 obtenemos lineas de puntos fijos.

```python
for a in [-2,3]:
  for b in [1,3,4]:
    A=np.array([[a,b],
              [b,a]])
    delta=np.linalg.det(A)
    traza=np.trace(A)
    print('para a=',a,'y b =',b,'Con traza =',traza,'y con determinante =',delta)
    X,Y=np.meshgrid(np.linspace(-2,2,12),np.linspace(-2,2,12))
    V= b*X + a*Y
    U= a*X + b*Y
    fig, ax = plt.subplots()
    Q= ax.quiver(X,Y,U,V,units ='width')

    def vdp(t,z):
      x,y=z 
      #Donde, j=y
      return[a*x+b*y,b*x+a*y] #Ecuaciones del sistema

    ti,tf=0,10
    t=np.linspace(ti,tf,500)
    for x0 in range(-2 , 3):
      for y0 in range(-2 , 3):
        for style in zip(styles):
          sol=solve_ivp(vdp,[ti,tf],[x0,y0],t_eval=t)
          plt.plot(sol.y[0],sol.y[1],style)
        plt.axis([-2,2,-2,2])
        plt.grid()
    ax.spines['left'].set_position(('data',0.0))
    ax.spines['bottom'].set_position(('data',0.0))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    plt.show()
```
